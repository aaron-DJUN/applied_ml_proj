**1. This notebook downloads images of movie posters from the movie dataset.** 

**2. The posters downloaded are subsequently used for a image classification problem.**

```python
# import libraries
import pandas as pd
import urllib
```

```python
# read movies
data_movies = pd.read_csv('../data/rotten_tomatoes_movies.csv')
poster_imgs = data_movies[['rotten_tomatoes_link', 'poster_image_url']]
N = len(poster_imgs)
```

```python
# download poster images
wrong_urls = []
for i in range(N):
    tomato_link = poster_imgs.rotten_tomatoes_link[i].replace('/','_')
    try:
        with open('posters/' + tomato_link + '.jpg','wb') as f:
            f.write(urllib.request.urlopen(poster_imgs.poster_image_url[i]).read())
        print('{}/{} is done.'.format(i+1, N))
    except Exception as e:
        wrong_urls.append(poster_imgs.id[i])
        print(e)
        continue
```

```python
# number of wrong urls
print('No. of Wrong Urls in data: {}'.format(len(wrong_urls)))
```
